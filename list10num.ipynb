```python
# Create a list of 10 numbers
lst = [23, 5, 67, 12, 89, 34, 2, 90, 45, 10]

# Sorting in descending order using simple logic (Bubble Sort)
n = len(lst)

for i in range(n):
    for j in range(0, n - i - 1):
        if lst[j] < lst[j + 1]:   # For descending order
            # Swap elements
            temp = lst[j]
            lst[j] = lst[j + 1]
            lst[j + 1] = temp

# Print sorted list
print("List in descending order:", lst)
```
