```python
# create a numpy array from the even fuctors of a geven number
import numpy as np

n = int(input("Enter number: "))

arr = []

for i in range(1, n+1):
    if n % i == 0 and i % 2 == 0:
        arr.append(i)

print(np.array(arr))
```
