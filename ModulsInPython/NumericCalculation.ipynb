```python
#NumericCalculation
#1.Function to add two nnumber
def calculateSum(x,y):
  return(x+y)

#2.Function to diff two number
def calculateSub(x,y):
  return(x-y)

#3.Function to Mul two number
def calculateMul(x,y):
  return(x*y)

#4.Function to divide two number
def calculateDiv(x,y):
  return(x/y)

#5.Function to Remender two number
def calculateRem(x,y):
  return(x%y)
```
