```python
# find second largest Number
nums = list(map(int, input("Enter numbers: ").split()))

largest = second = -999999  # initial small values

for i in nums:
    if i > largest:
        second = largest
        largest = i
    elif i > second and i != largest:
        second = i

print("Second Largest =", second)
```
