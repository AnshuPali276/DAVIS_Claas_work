```python
# Find Missing Numbers
nums = list(map(int, input("Enter numbers: ").split()))

n = len(nums) + 1  # expected size
total = n * (n + 1) // 2  # sum formula
# actual sum
actual = sum(nums)

print("Missing =", total - actual)
```
