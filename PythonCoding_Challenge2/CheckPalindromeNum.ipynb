```python
# Check Given Number Palindrome Or not
num = int(input("Enter number: "))
temp = num
rev = 0

while num > 0:
    rev = rev * 10 + num % 10
    num = num // 10

if temp == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```
