```python
# Rotate list by one position
nums = list(map(int, input("Enter numbers: ").split()))

# store last element
last = nums[-1]

# shift elements right
for i in range(len(nums)-1, 0, -1):
    nums[i] = nums[i-1]

# put last element at first
nums[0] = last

print(nums)
```
