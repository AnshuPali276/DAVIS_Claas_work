```python
# Common Elements Between two list
list1 = list(map(int, input("Enter list1: ").split()))
list2 = list(map(int, input("Enter list2: ").split()))

common = []

for i in list1:
    # check if element exists in both lists
    if i in list2 and i not in common:
        common.append(i)

print(common)
```
