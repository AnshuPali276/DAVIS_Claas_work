```python
# conver to Uppercase without built in funtion
text = input("Enter string: ")

result = ""

for ch in text:
    # check lowercase letters
    if 'a' <= ch <= 'z':
        # convert using ASCII
        result += chr(ord(ch) - 32)
    else:
        result += ch

print(result)
```
