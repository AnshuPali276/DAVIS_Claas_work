```python
# Find Students with highest Marks
students = {"A": 80, "B": 95, "C": 78}

max_marks = 0
topper = ""

for key in students:
    # check highest marks
    if students[key] > max_marks:
        max_marks = students[key]
        topper = key

print("Topper =", topper)
```
