```python
# Find duplicate character
text = input("Enter string: ")

duplicates = ""

for ch in text:
    # check if appears more than once and not already added
    if text.count(ch) > 1 and ch not in duplicates:
        duplicates += ch + " "

print("Duplicates =", duplicates)
```
