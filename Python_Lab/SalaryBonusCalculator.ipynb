```python
# Salary Bonus
def bonus(salary, years):
    if years > 10:
        return salary * 0.20
    elif years >= 5:
        return salary * 0.10
    else:
        return salary * 0.05

n = int(input("Enter number of employees: "))

for i in range(n):
    sal = float(input("Enter salary: "))
    exp = int(input("Enter years of experience: "))
    print("Bonus:", bonus(sal, exp))
```
