```python
import shapes3d

while True:
    print("\nSelect Shape:")
    print("1. Cube")
    print("2. Cuboid")
    print("3. Cylinder")
    print("4. Cone")
    print("5. Sphere")
    print("6. Exit")

    choice = int(input("Enter choice: "))

    if choice == 6:
        print("Exiting...")
        break

    print("\nSelect Operation:")
    print("1. Curved Surface Area")
    print("2. Total Surface Area")
    print("3. Volume")

    op = int(input("Enter operation: "))

    # Cube
    if choice == 1:
        a = float(input("Enter side: "))
        if op == 1:
            print("CSA =", shapes3d.cube_csa(a))
        elif op == 2:
            print("TSA =", shapes3d.cube_tsa(a))
        elif op == 3:
            print("Volume =", shapes3d.cube_volume(a))

    # Cuboid
    elif choice == 2:
        l = float(input("Enter length: "))
        b = float(input("Enter breadth: "))
        h = float(input("Enter height: "))
        if op == 1:
            print("CSA =", shapes3d.cuboid_csa(l, b, h))
        elif op == 2:
            print("TSA =", shapes3d.cuboid_tsa(l, b, h))
        elif op == 3:
            print("Volume =", shapes3d.cuboid_volume(l, b, h))

    # Cylinder
    elif choice == 3:
        r = float(input("Enter radius: "))
        h = float(input("Enter height: "))
        if op == 1:
            print("CSA =", shapes3d.cylinder_csa(r, h))
        elif op == 2:
            print("TSA =", shapes3d.cylinder_tsa(r, h))
        elif op == 3:
            print("Volume =", shapes3d.cylinder_volume(r, h))

    # Cone
    elif choice == 4:
        r = float(input("Enter radius: "))
        h = float(input("Enter height: "))
        l = float(input("Enter slant height: "))
        if op == 1:
            print("CSA =", shapes3d.cone_csa(r, l))
        elif op == 2:
            print("TSA =", shapes3d.cone_tsa(r, l))
        elif op == 3:
            print("Volume =", shapes3d.cone_volume(r, h))

    # Sphere
    elif choice == 5:
        r = float(input("Enter radius: "))
        if op == 1:
            print("CSA =", shapes3d.sphere_csa(r))
        elif op == 2:
            print("TSA =", shapes3d.sphere_tsa(r))
        elif op == 3:
            print("Volume =", shapes3d.sphere_volume(r))

    else:
        print("Invalid choice!")
```
