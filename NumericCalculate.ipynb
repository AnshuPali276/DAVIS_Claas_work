```python
#WAP to check a number is odd or even.
def check_even_odd(num):
  if num%2==0:
    print(num,"is even")
  else:
    print(num,"is odd")
number=int(input("Enter a number"))
# funtion calling
check_even_odd(number)
```

```python
#NumericCalculation
#1.Function to add two nnumber
def calculateSum(x,y):
  return(x+y)

#2.Function to diff two number
def calculateSub(x,y):
  return(x-y)

#3.Function to Mul two number
def calculateMul(x,y):
  return(x*y)

#4.Function to divide two number
def calculateDiv(x,y):
  return(x/y)

#5.Function to Remender two number
def calculateRem(x,y):
  return(x%y)
```
